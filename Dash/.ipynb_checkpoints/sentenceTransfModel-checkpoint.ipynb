```python
udemy_all_courses = pd.read_csv("coursera-courses.csv")#all the udemy courses
```

```python
udemy_all_courses.head()
```

```python
udemy_all_courses=udemy_all_courses.reset_index()
```

```python
features = ['course_name', 'description', 'course_difficulty']
for feature in features:
    udemy_all_courses[feature] = udemy_all_courses[feature].fillna('')
```

```python
def combined_features(row):
    return row['course_name']+" "+row['description']+" "+row['course_difficulty']
udemy_all_courses["combined_features"] = udemy_all_courses.apply(combined_features, axis =1)
```

```python
udemy_all_courses.head()
```

```python
sentences = udemy_all_courses['combined_features']
```

```python
sentences=sentences.to_list()
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model_name = 'bert-base-nli-mean-tokens'
```

```python
model = SentenceTransformer(model_name)
```
